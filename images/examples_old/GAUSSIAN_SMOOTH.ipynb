```python
import xcast as xc 
import xarray as xr 
import cartopy.crs as ccrs 

model = xr.open_dataset('data/MODEL_JJA.nc').prec
obs = xr.open_dataset('data/OBS_JJA.nc').precip.expand_dims({'M':[0]})
```

```python
pl = model.isel(time=0).plot(col='M', col_wrap=2, subplot_kws={'projection': ccrs.PlateCarree()})
for ax in pl.axs.flat:
    c = ax.coastlines()
```

```python
model = xc.gaussian_smooth(model) * 24
```
