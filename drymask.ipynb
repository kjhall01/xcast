```python
import src as xc 
import xarray as xr 
import cartopy.crs as ccrs 

obs = xr.open_dataset('data/OBS_JJA.nc').precip.expand_dims({'M':[0]})
```

```python
pl = obs.mean('time').plot(subplot_kws={'projection': ccrs.PlateCarree()} )
c = pl.axes.coastlines() 
```

```python
drymask = xc.drymask(obs, dry_threshold=30, quantile_threshold=0.3)
```

```python
pl = drymask.plot(subplot_kws={'projection': ccrs.PlateCarree()} )
c = pl.axes.coastlines()
```

```python
pl = (obs * drymask).mean('time').plot(subplot_kws={'projection': ccrs.PlateCarree()} )
c = pl.axes.coastlines()
```
